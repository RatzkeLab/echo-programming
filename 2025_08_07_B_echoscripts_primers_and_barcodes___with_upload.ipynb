```python
randseed = 250807
wells384 = np.array([r + str(c) for r in string.ascii_uppercase[:16] for c in range(1, 25)])
wells384_2d = np.reshape(np.array(wells384), [16,24])
fprimers_full = wells384_2d[::2,::2].flatten()
rprimers_full = wells384_2d[1::2,1::2].flatten()
fprimers = fprimers_full[12:]
rprimers = rprimers_full[12:]
print(fprimers)
print(rprimers)
```

```python
# prompt: read the columns "Well Position" and "Sequence" from /content/Plate Specs.xlsx into a pandas dataframe.
#df_barcodes = pd.read_excel('/content/Plate Specs.xlsx', usecols=['Well Position', 'Sequence Name', 'Sequence'])
df_barcodes = pd.read_excel('/content/drive/MyDrive/Sequencing + Processing/echoscripts_primers_sequencing_run/Plate Specs.xlsx', usecols=['Well Position', 'Sequence Name', 'Sequence'])
display(df_barcodes)
```

```python
# prompt: concatenate rprimers and fprimers into one list and then add that as a new column to df_barcodes

primers = np.concatenate((rprimers_full,fprimers_full))
df_barcodes['Storage'] = primers[:len(df_barcodes)]
display(df_barcodes)
```

```python
# prompt: create a new join table where df_transfer and df_barcodes are joined along the 'Source Well' column of df_transfer and the 'Storage' column of df_barcodes
# prompt: add a new column 'Direction' that contains the last non-numeric character from the string in 'Sequence Name' using the .str.extract method
df_joined = pd.merge(df_transfer, df_barcodes, left_on='Source Well', right_on='Storage')
# display(df_joined)
df_joined['Direction'] = df_joined['Sequence Name'].str.extract(r'([^\d])\d*$')
# display(df_joined)
df_joined = df_joined.drop(columns=['Source Well', 'Volume', 'Well Position', 'Sequence Name', 'Storage'])
df_joined['Sequence'] = df_joined['Sequence'].str.replace(' ', '')
display(df_joined)
```

```python
df_pivoted = df_joined.pivot(index='Destination Well', columns='Direction', values='Sequence').reset_index().reset_index(drop=True)#.set_index('Destination Well')

# df_pivoted.drop_index(inplace=True)
# df_pivoted.drop(['Direction'], axis=1, inplace=True)
# df_pivoted.rename(columns={'Destination Well':'SampleID','F':'FwIndex', 'R':'RvIndex'}, inplace=True)
# df_pivoted.drop
# df_pivoted.reset_index(inplace=True)
df_pivoted.columns = ['SampleID','FwIndex','RvIndex']
df_pivoted.set_index('SampleID', inplace=True)
# df_pivoted.droplevel(level=0)
# df_pivoted.reset_index(drop=True,inplace=True)
# display(df_pivoted)
df_pivoted.insert(1, 'FwPrimer', 'AGRGTTYGATYMTGGCTCAG')
df_pivoted.insert(3, 'RwPrimer', 'CGGYTACCTTGTTACGACTT')
display(df_pivoted)
# df_pivoted.to_csv("barcodes.txt", sep='\t')
```

```python
df_trimmed = df_pivoted.copy()
df_trimmed['FwIndex'] = df_trimmed['FwIndex'].str.slice(15,39)
df_trimmed['RvIndex'] = df_trimmed['RvIndex'].str.slice(15,39)
display(df_trimmed)
df_trimmed.to_csv(f"barcodes{randseed}.tsv", sep='\t')
```
